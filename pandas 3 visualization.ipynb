```python
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
### Drawa line in a diagram from position (0,12) to position (4,160)
xpoints = np.array([2,3])
ypoints = np.array([6,12])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([6,22])
ypoints = np.array([20,100])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([2,15])
ypoints = np.array([6,12])
plt.plot(xpoints,ypoints,"+")
plt.show()
```

```python
ypoints = np.array ([4,9,5,12])

plt.plot(ypoints,'*''r')
plt.show()
```

```python
ypoints = np.array([4,8,7,16])
xpoints= np.array([5,6,9,12])
plt.plot(xpoints,ypoints,marker = '*',ms = 30,color='k')
plt.show()
```

```python
ypoints = np.array([4,8,10])
plt.plot(ypoints,marker= 'o',ms = 40,mec='k',color='pink')
plt.show()
```

```python
ypoints = np.array([2,5,6,8,])
plt.plot(ypoints, marker ='o',ms = 30, mec ='b',mfc ='k')
plt.show()
```

```python
### Use both the mec and mfc arguments to color of the entire marker
ypoints = np.array([2,5,8,9])
plt.plot(ypoints, marker='o', ms= 20, mec ='y', mfc = 'r')
plt.show()
```

```python
ypoints = np.array([2,5,6,7])
plt.plot(ypoints, linestyle = 'dashed', marker='o', mec ='hotpink',ms=20,mfc='blue',linewidth=3)
plt.show()
```

```python
plt.plot(ypoints, linestyle='dotted', marker='o',ms=25,mfc="green",mec="red")
plt.show()
```

```python
xpoints=np.array([11,12,13])
ypoints=np.array([2,5,7])
plt.plot(xpoints,ypoints,marker="*",color="g",mfc="b",mec="r",ms=40,linestyle="dotted")
plt.show()
```

```python
### You can use the keyword argument linewidth or the shorter lw to change the width of the line. 

### The value is a floating number, in points
ypoints = np.array([2,4,6,8])
xpoints= np.array([1,3,5,7])
plt.plot(xpoints,ypoints , linewidth= 1, linestyle= "dotted", color="red", marker="o",ms=20,mec="g")
plt.show()
```

```python
y1 = np.array([3,4,9,12])
y2 = np.array([1,4,9,5,15])
plt.plot(y1,color="k",marker="*")
plt.plot(y2,color="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
x = np.array([70,80,95,105,120,135,140,145,150,160])
y = np.array([200,210,215,225,235,240,255,270,290,310])

plt.plot(x, y,color="r",linewidth=1)

plt.xlabel("Average Pulse",fontsize=10,color="red")
plt.ylabel("Caalorie Burnage",fontsize=10,color="g")
plt.title("sample",fontsize=30,color="blue")
plt.show()
```

```python
x = np.array([70,80,95,105,120,135,140,145,150,160])
y = np.array([200,210,215,225,235,240,255,270,290,310])

plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y,color="orange")

plt.grid(color="green",linestyle="dotted")

plt.show()
```

```python
### Set line properties for the Grid
### You can also set the line properties of the grid , like this: grid 
             ### (color = 'color',linestyle = 'linestyle', linewidth = number)
x = np.array([70,80,95,105,120,135,140,145,150,160])
y = np.array([200,210,215,225,235,240,255,270,290,310])

plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x,y)

plt.grid(color = 'hotpink', linestyle = '--', linewidth = 1.5)
plt.show()    
```

```python
# plot1:
x = np.array([1,2,3,4,])
y = np.array([3,5,8,9])

plt.subplot(1,2,1)
plt.plot(x,y)

#plot2:
x = np.array([0,2,4,6])
y = np.array([11,12,15,19])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
## The subplots() function takes three arguments that describes the layout of the figure.

## The layout is organized in rows and columns which are represented by the first and second arguments
```

```python
x = np.array([1,3,4,6])
y = np.array([2,5,8,9])

plt.subplot(2,3,1)
plt.plot(x,y)

x = np.array([0,2,3,4])
y = np.array([10,20,30,40])

plt.subplot(2,3,2)
plt.plot(x,y)

x = np.array([7,8,9,4])
y = np.array([4,5,6,3])

plt.subplot(2,3,3)
plt.plot(x,y)

x = np.array([0,1,4,7])
y = np.array([10,20,40,50])

plt.subplot(2,3,4)
plt.plot(x,y)
plt.show()
```
